```python
def rare_pok_city(listy):
    cities = []
    for i in listy : 
        cities.append(df[df['pokemonId'] == i ] ['city'])

    return cities
```

```python
fig = plt.figure(figsize=(15, 15))
map = Basemap()
map.fillcontinents(color='#191919',lake_color='#000000') # dark grey land, black lakes
map.drawmapboundary(fill_color='#000000')                # black background

map.drawcoastlines()
x,y = map(df.longitude.tolist(),df.latitude.tolist())
map.scatter(x,y,s=3,c='blue',zorder=5,lw=0)
```

```python
fig = plt.figure(figsize=(15, 15))
map = Basemap(projection='mill',
            llcrnrlat = 25,
            llcrnrlon = -130,
            urcrnrlat = 50,
            urcrnrlon = -60,
            resolution='l')
map.fillcontinents(color='#191919',lake_color='#000000') # dark grey land, black lakes
map.drawmapboundary(fill_color='#000000')                # black background
map.drawcoastlines()
map.drawcountries(linewidth=2)
map.drawstates(color='grey')
x,y = map(df.longitude.tolist(),df.latitude.tolist())
map.scatter(x,y,s=3,c='blue',zorder=5,lw=0)
```
